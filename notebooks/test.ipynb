```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
ranks = [4, 8, 16, 32]
multi_task_acc = [0.2, 42, 88, 95]
sort_asc = [0.4, 59, 86, 91]
sort_desc = [0.1, 28, 81, 91]
plus_one = [0.4, 40, 97, 99]
plus_two = [0, 42, 88, 95]
total_params = [36, 72, 144, 188]

# Plotting
plt.figure(figsize=(10, 6))

# Line plots for different metrics
plt.plot(ranks, multi_task_acc, marker='o', label='Multi-Task Acc (%)')
plt.plot(ranks, sort_asc, marker='s', label='Sort Asc (%)')
plt.plot(ranks, sort_desc, marker='^', label='Sort Desc (%)')
plt.plot(ranks, plus_one, marker='d', label='+1 (%)')
plt.plot(ranks, plus_two, marker='x', label='+2 (%)')

# Adding total params as secondary y-axis
ax1 = plt.gca()
ax2 = ax1.twinx()
ax2.plot(ranks, total_params, color='gray', linestyle='--', marker='o', label='~Total Params (k)')
ax2.set_ylabel('~Total Params (k)', color='gray')

# Title and labels
plt.title('LoRA Rank vs. Performance Metrics')
ax1.set_xlabel('LoRA Rank (r)')
ax1.set_ylabel('Performance Metrics (%)')
ax1.set_xticks(ranks)

# Legends for both y-axes
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='lower right')

# Grid and layout adjustments
plt.grid(alpha=0.4)
plt.tight_layout()

# Show the plot
plt.show()
```
